```python
import pandas as pd
import numpy as np
```

```python
gas = 'no'
df = pd.read_csv('GHG_Data/' + gas + '_by_country.csv', index_col=False)
df = df.drop(['Unnamed: 0'], axis=1)
df = df.rename(columns={'timestamp': 'Timestamp'})
```

```python
df.head()
```

```python
melted_df = df.melt(id_vars='Timestamp', var_name='Country Code', value_name=(gas.upper() + ' Levels'))
```

```python
df_split = melted_df.copy()
df_split['Year'] = pd.to_datetime(df_split['Timestamp']).dt.year
df_split['Month'] = pd.to_datetime(df_split['Timestamp']).dt.month
df_split = df_split.drop(['Timestamp'], axis=1)
df_split
```

```python
cols = df_split.columns.tolist()
cols = cols[-2:] + cols[:-2]
df_split = df_split[cols]
df_split
```

```python
# df_split.to_csv('GHG_Data/melted_' + gas + '_levels.csv')
```

```python
methane_df = pd.read_csv('GHG_Data/melted_methane_levels.csv')
co2_df = pd.read_csv('GHG_Data/melted_co2_levels.csv')
no_df = pd.read_csv('GHG_Data/melted_no_levels.csv')

methane_df = methane_df.drop(['Unnamed: 0'], axis=1)
co2_df = co2_df.drop(['Unnamed: 0'], axis=1)
no_df = no_df.drop(['Unnamed: 0'], axis=1)
```

```python
no_df
```

```python
merged_df = pd.merge(methane_df, co2_df, on=['Country Code', 'Month', 'Year'])
merged_df = pd.merge(merged_df, no_df, on=['Country Code', 'Month', 'Year'])
```

```python
merged_df = merged_df.rename(columns={'Country Code': 'CountryCode'})
merged_df
```

```python
merged_df.to_csv('GHG_Data/ghg_merged.csv')
```
